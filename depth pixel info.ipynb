```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd
```

```python
from PIL import Image                                                            
import numpy                                                                     
import glob

imageFolderPath = 'C:\\Users\\msi\\Desktop\\Depth\\png'
imagePath = glob.glob(imageFolderPath + '/*.png') 

im_array = numpy.array( [numpy.array(Image.open(img).convert('L'), 'f') for img in imagePath] )
```

```python
im_array
```

```python
data = im_array.flatten()
```

```python
freq=dict(pd.Series(data).value_counts())
```

```python
import seaborn as sns
sns.set_context("paper", rc={"font.size":8,"axes.titlesize":8,"axes.labelsize":5})   
fig, ax = plt.subplots(figsize=(500,100))

plot=sns.barplot(list(freq.keys()), list(freq.values()), ax=ax)
plot.tick_params(labelsize=5)
plt.show()
```
